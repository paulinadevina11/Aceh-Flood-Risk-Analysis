## Import Library

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("default")
```

## Load Dataset

```python
deforest_df = pd.read_csv(
    "Data Deforestasi Aceh - HAKA (Pemerintah Aceh)/Aceh 2 tree cover loss (GFW) - Subnational 2 tree cover loss.csv",
    header=0,
    engine="python"
)

deforest_df.head()
```

## Cek provinsi apa aja pada data

```python
deforest_df["subnational1"].unique()
```

## Cek Struktur Data

```python
deforest_df.info()
```

## Cek Kolom pada Data

```python
deforest_df.columns
```

### Kehilangan hutan di tiap provinsi (top 10)

```python
loss_cols = [c for c in deforest_df.columns if c.startswith("tc_loss_ha_")]

prov_deforest = (
    deforest_df
    .groupby("subnational1")[loss_cols]
    .sum()
    .sum(axis=1)
    .reset_index()
)

prov_deforest.columns = ["provinsi", "total_deforestasi_ha"]

prov_deforest = prov_deforest.sort_values(
    "total_deforestasi_ha", ascending=False
)

prov_deforest.head(10)
```

```python
top10 = prov_deforest.head(10)

plt.figure(figsize=(10,6))
plt.barh(
    top10["provinsi"][::-1],
    top10["total_deforestasi_ha"][::-1]
)
plt.xlabel("Total Kehilangan Hutan (ha)")
plt.title("Top 10 Provinsi dengan Deforestasi Terbesar di Indonesia\n(Global Forest Watch)")
plt.tight_layout()
plt.show()
```

### Tren Deforestasi di Indonesia

```python
long_df = deforest_df.melt(
    id_vars=["country", "subnational1", "subnational2"],
    value_vars=loss_cols,
    var_name="tahun",
    value_name="kehilangan_hutan_ha"
)

long_df["tahun"] = long_df["tahun"].str.extract(r"(\d{4})").astype(int)

indo_trend = (
    long_df
    .groupby("tahun")["kehilangan_hutan_ha"]
    .sum()
    .reset_index()
)

plt.figure(figsize=(10,5))
plt.plot(
    indo_trend["tahun"],
    indo_trend["kehilangan_hutan_ha"],
    marker="o"
)
plt.xlabel("Tahun")
plt.ylabel("Kehilangan Hutan (ha)")
plt.title("Tren Deforestasi Indonesia (2001â€“2024)\nGlobal Forest Watch")
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Aceh Peringkat ke-X Deforestasi

```python
prov_deforest["ranking"] = prov_deforest["total_deforestasi_ha"].rank(
    ascending=False, method="dense"
)

prov_deforest.loc[
    prov_deforest["provinsi"].str.contains("Aceh", case=False)
]
```

##### <span style="color:red;">Aceh termasuk peringkat 9 dalam deforestasi di Indonesia</span>

# Cleaning Data

## Pilih Threshold Hutan (Saya gunakan 30)

```python
deforest_df["threshold"].value_counts()
```

```python
deforest_df = deforest_df[deforest_df["threshold"] == 30].copy()
deforest_df.reset_index(drop=True, inplace=True)
```

## Pilih wilayah Aceh saja

```python
deforest_df = deforest_df[deforest_df["subnational1"] == "Aceh"].copy()
deforest_df.reset_index(drop=True, inplace=True)
```

```python
deforest_df["subnational2"].unique()
```

## Ubah nama kolom agar mudah dimengerti

```python
deforest_df = deforest_df.rename(columns={
    "subnational1": "provinsi",
    "subnational2": "kabupaten",
    "extent_2000_ha": "luas_hutan_2000_ha",
    "extent_2010_ha": "luas_hutan_2010_ha",
    "area_ha": "luas_wilayah_ha"
})

deforest_df["provinsi"] = deforest_df["provinsi"].str.title()
deforest_df["kabupaten"] = deforest_df["kabupaten"].str.title()

deforest_df.head()
```

```python
loss_cols = [c for c in deforest_df.columns if c.startswith("tc_loss_ha_")]

deforest_df = deforest_df.melt(
    id_vars=[
        "provinsi",
        "kabupaten",
        "luas_wilayah_ha",
        "luas_hutan_2000_ha",
        "luas_hutan_2010_ha"
    ],
    value_vars=loss_cols,
    var_name="tahun",
    value_name="kehilangan_hutan_ha"
)

deforest_df["tahun"] = deforest_df["tahun"].str.replace("tc_loss_ha_", "").astype(int)
deforest_df["sumber"] = "Global Forest Watch"
deforest_df.head(10)
```

# Analisis Deforestasi di Aceh

```python
deforest_df.info()
```

## Tren Deforestasi Aceh per tahun

```python
df_tahun = (
    deforest_df
    .groupby("tahun", as_index=False)["kehilangan_hutan_ha"]
    .sum()
)

plt.figure(figsize=(10,5))
plt.plot(
    df_tahun["tahun"],
    df_tahun["kehilangan_hutan_ha"],
    marker="o"
)
plt.title("Tren Kehilangan Hutan di Aceh per Tahun")
plt.xlabel("Tahun")
plt.ylabel("Kehilangan Hutan (ha)")
plt.grid(True)
plt.show()
```

##### <span style="color:red;">Dari gambar tersebut, terlihat bahwa kehilangan hutan terbesar di seluruh wilayah Aceh adalah pada tahun 2010. Grafik tidak terlalu stabil turun ataupun naik, namun **terlihat ada peningkatan kembali pada 2 atau 3 tahun terakhir sebelum 2025**</span>

## Normalisasi

### Untuk mengetahui kabupaten mana paling parah

```python
deforest_df["deforestasi_persen_wilayah"] = (
    deforest_df["kehilangan_hutan_ha"] /
    deforest_df["luas_wilayah_ha"]
) * 100

df_norm = (
    deforest_df
    .groupby("kabupaten", as_index=False)["deforestasi_persen_wilayah"]
    .mean()
    .sort_values("deforestasi_persen_wilayah", ascending=False)
)

df_norm.head(100)
```

# Perbandingan Deforestasi Aceh per Kabupaten

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load Data
# Pastikan nama file sesuai dengan yang Anda miliki
file_path = 'Data Deforestasi Aceh - HAKA (Pemerintah Aceh)/Aceh 2 tree cover loss (GFW) - Subnational 2 tree cover loss.csv'
df = pd.read_csv(file_path)

# 2. Cleaning & Filtering
# - Ambil hanya Provinsi Aceh
# - Ambil threshold 30 (Standar umum untuk definisi hutan di Indonesia/tropis)
df_aceh = df[(df['subnational1'] == 'Aceh') & (df['threshold'] == 30)].copy()

# 3. Reshaping Data (Melting)
# Mengubah kolom 'tc_loss_ha_2001' s/d '2024' menjadi baris ke bawah
loss_cols = [col for col in df_aceh.columns if 'tc_loss_ha_' in col]

df_melt = df_aceh.melt(
    id_vars=['subnational2'],      # Identitas (Nama Kabupaten)
    value_vars=loss_cols,          # Kolom tahun yang akan dicairkan
    var_name='tahun_raw',          # Nama kolom baru untuk tahun mentah
    value_name='luas_deforestasi'  # Nama kolom baru untuk nilai luas
)

# 4. Membersihkan Kolom Tahun
# Menghapus teks 'tc_loss_ha_' dan mengubah menjadi angka integer
df_melt['tahun'] = df_melt['tahun_raw'].str.replace('tc_loss_ha_', '').astype(int)

# 5. Membuat Visualisasi (Spaghetti Plot)
plt.figure(figsize=(14, 8))

# Menggunakan seaborn lineplot dengan 'hue' untuk membedakan kabupaten
sns.lineplot(
    data=df_melt,
    x='tahun',
    y='luas_deforestasi',
    hue='subnational2',    # Membedakan warna tiap kabupaten
    palette='tab20',       # Palet warna yang kontras untuk banyak kategori
    linewidth=2,
    alpha=0.8,             # Sedikit transparan agar garis yang bertumpuk terlihat
    marker='o'             # Menambahkan titik pada setiap tahun
)

# Kustomisasi Grafik
plt.title('Tren Deforestasi per Kabupaten di Aceh (2001-2024)', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Tahun', fontsize=12)
plt.ylabel('Luas Deforestasi (Ha)', fontsize=12)
plt.xticks(range(2001, 2025, 1), rotation=45) # Menampilkan semua tahun
plt.grid(True, linestyle='--', alpha=0.5)

# Mengatur Legenda di luar grafik agar rapi
plt.legend(
    title='Kabupaten/Kota',
    bbox_to_anchor=(1.02, 1),
    loc='upper left',
    borderaxespad=0,
    ncol=1,
    fontsize='small'
)

plt.tight_layout()
plt.show()
```
